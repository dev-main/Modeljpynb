```python
from mpl_toolkits import mplot3d
 
import numpy as np
import matplotlib.pyplot as plt
 
import pandas
 
fig = plt.figure()
ax = plt.axes(projection="3d")
 
df = pandas.DataFrame({
   'prop': ['change', 'scaling', 'buildc', 'operc'],
   'col2': [1, 3, 5, 6],
   'col3': [1, 3, 5, 6],
   'col4': [1, 3, 5, 6]
})
 
ax.set_xlabel('xxxxx')
ax.set_ylabel('y1')
ax.set_zlabel('z')
 
ax.scatter3D(df.col2, df.col3, df.col4, c=df.col4, cmap='hsv');
 
ax.text(1, 1, 1, '(1)' , size=12, zorder=1, color='k')
ax.text(3,3, 3, '(2)' , size=12, zorder=1, color='k')
ax.text(5, 5, 5, '(3)' , size=12, zorder=1, color='k')
ax.text(6, 6, 6, '(4)' , size=12, zorder=1, color='k')
 
ax.bar3d(df.col2, df.col3, df.col4, 0.1, 0.1, -df.col4, color='g')
plt.show()
```
